```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1997_1998=pd.read_csv('data_1997_1998.csv')
df_evaluation=pd.read_csv('evaluation.csv')
df_updated=pd.read_csv('updated.csv')
```

```python
df_evaluation.drop(columns='Unnamed: 0',inplace=True)
df_evaluation.head()
```

```python
df_1997_1998.drop(columns='Index',inplace=True)
```

```python
datasets=[df_updated,df_evaluation,df_1997_1998]
features=df_1997_1998.columns
n_rows=len(features)
print(features)
```

```python
fig,axes=plt.subplots(n_rows,3,figsize=(15,5*n_rows))

for i,feature in enumerate(features):
    for j,dataset in enumerate(datasets):
        sns.histplot(x=dataset[feature],kde=True,ax=axes[i,j])
        axes[i,j].set_title(f'{j+1} - {feature}')

fig.suptitle('1-original\n2-evaluation\n3-1997_1998')

plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```
